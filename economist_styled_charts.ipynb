```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(facecolor='white', figsize=(12,8))

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Calibri']

#data
labels=['USA', 'China', 'Russia', 'Italy','Germany','Spain','Belgium','India','Singapur','Canada']
sizes=[30,35,40,45,20,18,19,5,6,7]
colors=['#00485d','#00485d','#00485d','#00485d','#00a1ce','#00a1ce','#00a1ce','#86d7f2','#86d7f2','#86d7f2']
 
#расстояния между секторами
explode = [0.05]*len(labels)
 
kw = dict(arrowprops=dict(arrowstyle="-"),
          zorder=0, va="center")
wedges, texts = ax.pie(sizes, colors = colors, startangle=80, explode = explode)
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1 #расстояние между лейблами
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(labels[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.1*y),
                horizontalalignment=horizontalalignment, **kw,fontsize=16)

#draw circle
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

#text
ax.text(-1.5,1.6,'Military spending',fontdict={'fontweight' : 600,'fontsize':22})
ax.text(-1.5,1.3,'Characteristics of dogs registered with the UK\'s \nKennel Club, average when fully grown',fontdict={'fontsize':18})

ax.text(0,0,'Total: \n$1.75trn',fontdict={'fontsize':18},ha='center')
ax.text(0,-0.1,'(-45.34%)',fontdict={'fontsize':18},ha='center',color='#ef614b')

ax.text(-1.5,-1.3,'Source: Stockholm International Peace Research Institute',color='grey',fontdict={'fontsize':15})

# Equal aspect ratio ensures that pie is drawn as a circle  
plt.tight_layout()
plt.show()
```
