```python
%matplotlib inline

import string
import random
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(facecolor='#cddee6')
plt.rcParams['font.family'] = 'serif'

# Example data
country = ('Japan*', 'Italy', 'France', 'G7', 'Britain','United States','Germany','Canada')
x_pos = [121,115,86,80,75,70,50,37]
y_pos=np.arange(len(country))

# Remove axis lines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#axis to the top
ax.xaxis.tick_top()

my_colors = ('#04526d','#04526d','#04526d','#04526d','#04526d','#04526d','#04526d','#00a5da')

ax.barh(y_pos,x_pos, align='center',height=0.6,color=my_colors)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_yticks(y_pos)
ax.set_yticklabels(country)

#axos label font size
ax.tick_params(axis='both', which='major', labelsize=12)

#Titles
ax.set_title(label='Trudeau\'s trousseau \nNet government debt as %of GDP',
             loc='left',fontdict={'fontweight' : 600,'fontsize':12})
ax.text(-22,8.2,'Source: IMF')


#Background
ax.set_facecolor('#cddee6')

#грид линии
plt.grid(axis='x', color='white',linewidth=2)
ax.set_axisbelow(True) #грид под графиком

#черточки на осях
ax.tick_params(direction='inout',length=0) 



plt.show()
```
