```python
%matplotlib inline

import string
import random
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(facecolor='black')
plt.rcParams['font.family'] = 'serif'

# Example data
country = ('Japan*', 'Italy', 'France', 'G7', 'Britain','United States','Germany','Canada')
x_pos = [56,43,86,68,75,55,50,120]
y_pos=np.arange(len(country))

# Remove axis lines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)


bars=ax.barh(y_pos,x_pos, align='center',height=0.6,color="#fe5900")
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_yticks(y_pos)
ax.set_yticklabels(country,color="white")

#    
ax.tick_params(axis='y', which='major', pad=15)

#ax.tick_params(axis='x', colors='white')


#axos label font size
ax.tick_params(axis='both', which='major', labelsize=12)


#Background
ax.set_facecolor('black')


#  
ax.tick_params(direction='inout',length=0) 

#  
c=0
for i in x_pos:
    ax.text(i+2,c+0.1,i,fontdict={'fontsize':12},color='white')
    c+=1

    
#text
ax.text(-35,-1.8,'There\'s a Chance',fontdict={'fontweight' : 600,'fontsize':22},color='white')
ax.text(-35,-1,'Characteristics of dogs registered with the UK\'s',fontdict={'fontsize':18},color='white')

ax.text(-35,8,'Source: Stockholm International Peace Research Institute',color='#d5d5d5',fontdict={'fontsize':12})
plt.savefig('example1.svg', dpi=300, facecolor=fig.get_facecolor(), edgecolor='none',bbox_inches='tight')
plt.show()
```
